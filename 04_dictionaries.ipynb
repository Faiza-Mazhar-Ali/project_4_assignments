```python
#00_count_nums.md
def get_user_numbers():
    """
    Ask the user to enter numbers and store them in a list.
    Stop asking for input when the user enters a blank line.
    """
    user_numbers = []
    while True:
        user_input = input("Enter a number: ")

        if user_input == "":  # Stop when the user presses Enter
            break

        num = int(user_input)  # Convert to integer
        user_numbers.append(num)

    return user_numbers

def count_nums(num_lst):
    """
    Count how many times each number appears in the list.
    Return a dictionary with number-frequency pairs.
    """
    num_dict = {}
    for num in num_lst:
        num_dict[num] = num_dict.get(num, 0) + 1  # Update count using .get()

    return num_dict

def print_counts(num_dict):
    """
    Print out each number and how many times it appears.
    """
    for num, count in num_dict.items():
        print(f"{num} appears {count} times.")  # Using f-string for cleaner output

def main():
    """
    Get user input, count occurrences, and print results.
    """
    user_numbers = get_user_numbers()
    num_dict = count_nums(user_numbers)
    print_counts(num_dict)

# Required Python boilerplate
if __name__ == '__main__':
    main()
```

```python
#01_phonebook.md
def add_contacts():
    """
    Collects names and phone numbers from the user and stores them in a dictionary.
    Returns the phonebook.
    """
    contacts = {}  # Empty dictionary to store phonebook

    while True:
        name = input("Enter name (or press Enter to stop): ").strip()
        if not name:
            break  # Stop input when user presses Enter

        number = input(f"Enter phone number for {name}: ").strip()
        contacts[name] = number  # Store name-number pair

    return contacts


def display_contacts(phonebook):
    """
    Displays all contacts stored in the phonebook.
    """
    print("\nüìû Phonebook Contacts:")
    if not phonebook:
        print("No contacts found.")
    else:
        for name, number in phonebook.items():
            print(f"{name} ‚Üí {number}")


def search_contact(phonebook):
    """
    Allows user to search for a contact's number by name.
    """
    while True:
        name = input("\nüîç Enter name to look up (or press Enter to exit): ").strip()
        if not name:
            break  # Stop searching when user presses Enter

        if name in phonebook:
            print(f"{name}'s number: {phonebook[name]}")
        else:
            print(f"{name} is not in the phonebook.")


def main():
    """
    Runs the phonebook program.
    """
    phonebook = add_contacts()
    display_contacts(phonebook)
    search_contact(phonebook)


if __name__ == "__main__":
    main()
```

```python
#02_pop_up_shop.md
def main():
    # Dictionary containing fruits and their prices
    fruits = {
        'apple': 1.5,
        'durian': 50,
        'jackfruit': 80,
        'kiwi': 1,
        'rambutan': 1.5,
        'mango': 5
    }

    total_cost = 0  # Variable to store the total cost

    for fruit, price in fruits.items():
        while True:  # Keep asking until a valid number is entered
            try:
                quantity = int(input(f"How many ({fruit}) do you want?: "))
                if quantity < 0:
                    print("Please enter a valid non-negative number!")
                else:
                    total_cost += price * quantity
                    break  # Exit the loop if input is valid
            except ValueError:
                print("Invalid input! Please enter a number.")

    # Print total cost formatted to two decimal places
    print(f"Your total is ${total_cost:.2f}")


# Required Python boilerplate
if __name__ == '__main__':
    main()
```

```python
#03_powerful_passwords.md
import hashlib

def verify_login(email, users_db, password_attempt):
    """
    Checks if the provided password matches the stored hashed password.

    Parameters:
        email (str): The email of the user attempting to log in.
        users_db (dict): Dictionary storing emails as keys and hashed passwords as values.
        password_attempt (str): The password entered by the user.

    Returns:
        bool: True if the password is correct, False otherwise.
    """
    hashed_attempt = generate_hash(password_attempt)
    return users_db.get(email) == hashed_attempt

def generate_hash(password):
    """
    Converts a password into a SHA256 hash.

    Parameters:
        password (str): The password to hash.

    Returns:
        str: The hashed password.
    """
    return hashlib.sha256(password.encode()).hexdigest()

def main():
    # Database of stored emails and their hashed passwords
    stored_users = {
        "user1@example.com": generate_hash("securepass123"),
        "admin@website.org": generate_hash("admin!@#2024"),
        "guest@demo.net": generate_hash("guestpass")
    }

    # Testing login attempts
    print(verify_login("user1@example.com", stored_users, "securepass123"))  # Expected: True
    print(verify_login("admin@website.org", stored_users, "wrongpassword"))  # Expected: False
    print(verify_login("guest@demo.net", stored_users, "guestpass"))  # Expected: True
    print(verify_login("unknown@fake.com", stored_users, "randompass"))  # Expected: False

if __name__ == "__main__":
    main()
```
